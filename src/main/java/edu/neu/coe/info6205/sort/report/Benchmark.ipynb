```
import matplotlib.pyplot as plt
import numpy as np
import math
```

```
N= [250, 500, 1000, 2000, 4000, 8000, 16000]

random= [0.08, 0.21, 0.77, 2.59, 10.37 ,46.32 ,189.51]
ordered= [0,0,0,0,0.01,0.02,0.04]
pordered=[0.05,0.16,0.64,2.22,9.29,36.24,144.89]
rordered=[0.1,0.49,1.91,6.04,23.72,95.34,446.76]
```

```
plt.plot(N, random, label = "Random ordered array")
plt.plot(N, ordered, label = "Ordered array")
plt.plot(N, pordered, label = "Partially ordered array")
plt.plot(N, rordered, label = "Reverse ordered array")
  
# naming the x axis
plt.xlabel('Run time')
# naming the y axis
plt.ylabel('N')
# giving a title to my graph
plt.title('Insertion sort')
  
# show a legend on the plot
plt.legend()
  
# function to show the plot
plt.show()
```
