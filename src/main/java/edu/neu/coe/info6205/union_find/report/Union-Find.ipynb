```
import matplotlib.pyplot as plt
import numpy as np
import math
```

```
N= [100,200,400,800,1600,3200,6400,12800,25600,51200,102400,204800,409600]

M= [254.95, 603.52, 1359.52, 2908.04, 6304.13, 13525.99, 30423.56, 63250.19, 134399.66, 298967.24, 620226.9, 1321211.3, 2770013.64]

NlogN= [ 0.5*n*math.log(n) for n in N]

print(M)
print(NlogN)
```

```
plt.plot(N, M, label="Experiment")
plt.plot(N, NlogN, label="Half NlogN")

# naming the x axis
plt.xlabel('Number of pairs')
# naming the y axis
plt.ylabel('No of objects')
# giving a title to my graph
plt.title('Uninon Find')
  
# show a legend on the plot
plt.legend()
  
# function to show the plot
plt.show()
```
